```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.imdb.com/chart/top?sort=rk,asc&mode=simple&page=1"
r = requests.get(url)
```

```python
soup = bs(r.content)
print(soup.prettify())
# Remove square bracket can use .join()
```

```python
table = soup.find(class_ = "titleColumn")
title = table.find('a')
print(title)
```

```python
title.get_text()
```

```python
table_all = soup.find_all(class_ = "titleColumn")
print(table_all)
```

```python
titles = []
for i in range(0,len(table_all)):
    num = table_all[i].find('a').get_text()
    titles.append(num)


print(titles)
```

## Years Column

```python
year= soup.find_all(class_ = "secondaryInfo")
print(year)
```

```python
year[1]
```

```python
year[0].get_text()
```

```python
years = []
for i in range(0,len(year)):
    num = year[i].get_text()
    years.append(num)

print(years)

```

## Rating Column

```python
rating = soup.find_all(class_ = "ratingColumn imdbRating")
print(rating)
```

```python
rating[0].get_text()
```

```python
ratings = []
for i in range(0,len(rating)):
    num = rating[i].get_text()
    ratings.append(num)
```

```python
ratings
```

## Final draft

```python
import pandas as pd
df = pd.DataFrame(zip(titles,years,ratings), columns = ['title','year','rating'])
df.head()
```

```python
import re 
def cleanText(text):
    text = re.sub('\\n', '',text)
    text = re.sub('\(', '',text)
    text = re.sub('\)', '',text)
    return text
```

```python
df['year'] = df['year'].apply(cleanText)
df['rating'] = df['rating'].apply(cleanText)
df.head()
```

```python
# Adding a rank column
df.insert(0, 'rank', list(range(1,251)))
```

```python
df.info()
```

```python
# Transforming data into correct type
df['year'] = df['year'].astype(int)
df['rating'] = df['rating'].astype(float)
df['rank'] = df['rank'].astype(int)
```

```python
df.info()
```

```python
df['year'].max()
```

```python
# Frequency table
frequency_table = df.copy()
frequency_table['freq'] = pd.cut(df['year'],10)
# Count of each interval
freq_table = frequency_table.groupby("freq")['rank'].count().reset_index().rename(columns={'rank': 'count'})

# Sorting descending
freq_table.sort_values(['count'], ascending=False) 
```

```python
import matplotlib.pyplot as plt

plt.hist(x = df['year'],bins= 10, color = '#0504aa', alpha = 0.7,rwidth=0.85)
```

## Revenue of the movies

```python
url = "https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW"
r = requests.get(url)
```

```python
soup = bs(r.content)
print(soup.prettify())
# Remove square bracket can use .join()
```

```python
table = soup.find_all(class_ = "a-text-left mojo-field-type-title")
```

```python
movies = []
for i in range(0,len(table)):
    num = table[i].get_text()
    movies.append(num)


print(movies)
```

```python
revenue = soup.find_all(class_ = "a-text-right mojo-field-type-money")
```

```python
revenues = []
for i in range(0,len(revenue)):
    num = revenue[i].get_text()
    revenues.append(num)


print(revenues)
```

```python
year = soup.find_all(class_ = "a-text-left mojo-field-type-year")
```

```python
years = []
for i in range(0,len(year)):
    num = year[i].get_text()
    years.append(num)


print(years)
```

### Transforming into dataframe

```python
rev = pd.DataFrame(zip(movies,years,revenues), columns = ['movie','year','revenue'])
rev.head()
```

```python
# Updating the cleanText function by adding the last two column
# These two columns can remove the comma and the dollar sign.
# After removing comma and $, we can change the data type into integer for rev['revenue']
def cleanText(text):
    text = re.sub('\\n', '',text)
    text = re.sub('\(', '',text)
    text = re.sub('\)', '',text)
    text = re.sub('\,', '',text)
    text = re.sub('\$', '',text)
    return text
```

```python
rev['revenue'] = rev['revenue'].apply(cleanText)
```

```python
# Transforming data into correct type
rev['movie'] = rev['movie'].astype(str)
rev['year'] =rev['year'].astype(int)
rev['revenue'] = rev['revenue'].astype(int)
```

```python
rev.head()
```

```python
# As we only have 200 rows on revenue table for a url, so we would limit the other table rows into 200
# In order to make it consistent
df_200 = df[0:200]
```

```python
# Doing a inner join to merge two table on title and movie column
# There is a duplicate movie,"The Lion King" in df_join. We would drop it later
df_join = pd.merge(df_200,rev,left_on='title',right_on='movie',how = 'inner').drop(['title','year_x'],axis = 1)
df_join.rename(columns = {'year_y':'year'}, inplace= True)
df_join
```

```python
# Exporting df_join into csv file
df_join.to_csv('df_join.csv')
```

```python
# Moving rating column into the third position
df_join.insert(3,'rating',df_join.pop('rating'))
```

```python
# Droping the duplicate movie, 'The Lion King'
df_join = df_join.drop(labels = 7,axis=0)
df_join
```

```python
# Here we can see that if we extract the top 200 from each table, there are only 23 of them are in both table.
df_join.sort_values('rank')
```
