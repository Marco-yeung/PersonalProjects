```python
import pandas as pd
import os
import plotly.express as px
from datetime import datetime
from plotly.subplots import make_subplots

```

```python
# present working directory
pwd = os.getcwd()
pwd
```

```python
df = pd.read_csv(pwd + "/NFT_Sales.csv")
```

```python
df.head
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
# Changing column date into datetime object
df['Date'] = pd.to_datetime(df["Date"])
```

```python
# Changing the column name to make it more clear
df.rename(columns = {'Primary_Sales':'Number_of_Primary_Sales',
                    'Secondary_Sales_cumsum': 'Secondary_Sales'
                    },
                    inplace = True)
```

```python
# Calculating the average sales per transaction each day
df['AverageUSD_per_Trans'] = df['Sales_USD']/df['Number_of_Sales']
df['AverageUSD_per_Trans'] = df['AverageUSD_per_Trans'].round(2)
```

### Number column describe

```python
df.describe()
# AverageUSD_cum = Sales_USD_cumsum/Number_of_Sales_cumsum
# Active market wallets is defined as the number of unique wallets which bought or sold an asset.
```

### Date Analysis

```python
# plotly method
fig1 = px.line(df, x = 'Date', y = 'Active_Market_Wallets',
        title = 'Number of Active Market Wallets over Time',
        labels = {
                'Active_Market_Wallets': 'Number of Active Market Wallets'
        }
        )
```

```python
fig2 = px.line(df, x = 'Date', y = 'Sales_USD',
        title = 'Amount of Sales per Day (USD)',
        labels = {'Sales_USD' : 'Sales (USD)'}
        )
```

```python
fig3 = px.line(df, x = 'Date', y = 'Number_of_Sales',
        title = 'Number of Sales per Day',
        labels = {'Number_of_Sales' : 'Number of Sales'})
```

```python
fig4 = px.line(df, x = 'Date', y = 'Number_of_Primary_Sales',
        title = 'Number of Primary Sales per Day',
        labels = {'Number_of_Primary_Sales' : 'Number of Primary Sales'})
```

```python
fig5 = px.line(df, x = 'Date', y = 'Secondary_Sales',
        title = 'Amount of Secondary Sales per Day (USD)',
        labels = {'Secondary_Sales' : 'Secondary Sales (USD)'})
```

```python
fig7 = px.line(df, x = 'Date', y = 'Sales_USD_cumsum',
title = 'Cumulative Sum of Sales (USD)',
        labels = {'Sales_USD_cumsum' : 'Cumulative Sum of Sales (USD)'
        })
```

```python
fig8 = px.line(df, x = 'Date', y = 'AverageUSD_cum',
        title = 'Daily Average Cummulative Sales (USD)',
        labels = {'AverageUSD_cum' : 'Cummulative Average Sales (USD)'})
```

```python
fig9 = px.line(df, x = 'Date', y = 'AverageUSD_per_Trans',
        title = 'Daily Average Sales (USD)',
        labels = {'AverageUSD_per_Trans' : 'Average Sales (USD)'})
```

```python
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Number of Active Market Wallets over Time',
                    'Amount of Sales per Day (USD)',
                    'Number of Sales per Day',
                    'Daily Average Sales (USD)'))
```

```python
fig.add_trace(fig1['data'][0], row=1, col=1)
fig.add_trace(fig2['data'][0], row=1, col=2)
fig.add_trace(fig3['data'][0], row=2, col=1)
fig.add_trace(fig9['data'][0], row=2, col=2)
fig.update_layout(height=700)
fig.show()


# 1st plot
# there is a spike at 10-10-21 (only last for a day)
# An early spike around the start of Dec2017, then drop to normal level for the following three years
# after that there is 2 days over 10000, and a day close to 10000

# 2nd plot
# all spike found at 2021
# spike 03-05-21 (104.6677M)
# a large spike at 29-08-21(331M), 2nd large spike at 23-08-21(243M)
# the mean level after Sep is rised. Although after two large spikes

#3rd plot
# small spike at 09-12-17, then slightly drop to normal level
# greater variance from Aug and higher mean level
# a more volatile period at 17-10-21


# 4th plot
# the mean level daily avg sales have a drastic increase since 2021
# the variance is a lot larger than before too
# a short period(2 month) avg sales is pretty constant (21May-21-July), maybe a long period as it happen between 1 year (***BUT WHY?***)


```

### Key Findings

Mostly the NFT boom are started after 2021.

For the plot of active market wallets,
there is a spike at 10-10-21. (only last for a day)\
After that there is 2 days over 10000, and a day close to 10000, which is 1/3 of the peak.

Maybe some relationship between Active market wallets and no. of sales per day. As there is a peak for Active market wallets, 10 days later, there has the peak for no. of sales per day.

number of sales seems don't have a relationship with sales amount. As the peak level is completely reversed. 

Daily average sales is having a decreasing trend past 2 month.

The number of sales, number of primary sales and the amount of secondary sales shows the variance increased over time since 2021.

From the plot of Cumulative active market wallets, the number of active market wallets are increasing constantly from 2018 to 2021. However, starting from 2021, the number of wallets are rising significantly. The number of wallets at the end of 2021 is around 4 times the number at the start of 2021.
